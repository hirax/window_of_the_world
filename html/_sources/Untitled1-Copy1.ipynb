```python
import osmnx as ox
import networkx as nx
import folium

# 2点の緯度経度を指定します
point1 = (35.6895, 139.6917)  # 東京駅付近
point2 = (35.6581, 139.7414)  # 東京タワー付近

# 指定した地点を含むグラフを取得します
G = ox.graph_from_point(point1, dist=2000, network_type='walk')

# 2点に最も近いノードを取得します
orig_node = ox.distance.nearest_nodes(G, point1[1], point1[0])
dest_node = ox.distance.nearest_nodes(G, point2[1], point2[0])

# 最短経路を計算します
shortest_route = nx.shortest_path(G, orig_node, dest_node, weight='length')

# 最短経路の座標を取得します
route_coords = [(G.nodes[node]['y'], G.nodes[node]['x']) for node in shortest_route]

# Foliumマップを作成し、2点間の最短経路をプロットします
m = folium.Map(location=point1, zoom_start=14)

# 2点をマーカーで表示します
folium.Marker(location=point1, popup="Start", icon=folium.Icon(color='green')).add_to(m)
folium.Marker(location=point2, popup="End", icon=folium.Icon(color='red')).add_to(m)

# 最短経路をラインで表示します
folium.PolyLine(route_coords, color="blue", weight=2.5, opacity=1).add_to(m)

# 地図を保存または表示します
#m.save('shortest_route_map.html')
m  # Jupyter Notebook上で表示する場合
```
